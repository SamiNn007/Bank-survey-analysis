```python
# Contingency tables and independence of effects

## H0: age group independent of the areas of imporovement ---
## H1: others were not easily influenced by the header of the survey

#we just need the observed values 
## cars_table=[[256,74],[41,42],[66,34]]

# significance level 
alpha = 0.05

st.chi2_contingency(np.array(data_crosstab))
#statistic, p-value, degrees of freedom, expected probs if H0 holds
```
